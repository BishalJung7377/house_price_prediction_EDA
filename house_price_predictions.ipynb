### Missing Values


```python
features_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes == 'O']
features_nan
```

```python
for feat in features_nan:
    print(f'{feat} is missing {np.round(df[feat].isnull().mean(),4)} % of values')
```

```python
## replacing the missing values
def replace_cat_features(df, features_nan):
    data = df.copy()
    data[features_nan] = data[features_nan].fillna("Missing")
    return data

df = replace_cat_features(df, features_nan)
df[features_nan].isnull().sum()
```

```python
df.head()
```

```python
### checkiing percentage of missing numerical values
numerical_with_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtypes != 'O']
for feat in numerical_with_nan:
    print(f'{feat} has {np.round(df[feat].isnull().mean(),4)} % of values')
```

```python
## replacing the missing numerical values
for feature in numerical_with_nan:
    median_value = df[feature].median()
    
    df[feature + 'nan'] = np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value, inplace = True)
    
df[numerical_with_nan].isnull().sum()
    
```

```python
df.head()
```

#### date time variables are called temporal variable

```python
## handling temporal values
for feat in ['yearbuilt','yearremodadd','garageyrblt']:
    df[feat] = df["yrsold"] - df[feat]
```

### handling rare categorical features
##### we will remove categorical variables that are present less than 1% of the observations

```python
## seperating categorical features
cat_features = [feature for feature in df.columns if df[feature].dtype =='O']
```

```python
cat_features
```

```python

for feature in cat_features:
    temp = df.groupby(feature)['saleprice'].count()/len(df)
    temp_df = temp[temp > 0.01].index
    df[feature]= np.where(df[feature].isin(temp_df), df[feature],'rare_var')
```

```python
for feature in categorical_features:
    labels_ordered = df.groupby(feature)["saleprice"].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,0)}
    df[feature] = df[feature].map(labels_ordered)
```
